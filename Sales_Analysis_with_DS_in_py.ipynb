```python
all_data = pd.read_csv("all_data.csv")
all_data.head()
```

cleaning up the data

```python
all_data.head(50)
```

drop rows of NaN

```python
nan_df = all_data[all_data.isnull().any(axis=1)]
nan_df.head()

all_data = all_data.dropna(how='all')
all_data.head()
```

### Find 'Or' and deleting it


```python
all_data = all_data[all_data['Order Date'].str[0:2] != 'Or']
all_data.head()
```


#### converting columns to the correct type

```python
all_data['Quantity Ordered'] = pd.to_numeric(all_data['Quantity Ordered'])
all_data['Price Each'] = pd.to_numeric(all_data['Price Each'])
```

# Augment data with additional coloumns  

adding month column

```python
all_data["Month"]= all_data["Order Date"].str[0:2]
all_data["Month"] = all_data["Month"].astype('int32')
all_data.head()
```

#### Adding a sales column 

```python
all_data['Sales'] = all_data['Quantity Ordered'] * all_data['Price Each']
all_data.head()
```

#### Adding city column 

```python
# using the .apply() method
```

```python
def get_city(address):
    return address.split(',')[1]

def get_state(address):
    return address.split(',')[2].split(' ')[1]

all_data['States'] = all_data['Purchase Address'].apply(lambda x: get_city(x) + ' ('+ get_state(x)+ ') ')
# or use f strings
all_data.head()
```

## Best Month for Sales and how much was earned that month.

```python
results = all_data.groupby('Month').sum()
```

#### What City has the highest number of sales?


```python
results = all_data.groupby('States').sum()
```

```python
import matplotlib.pyplot as plt

States = [state for state, df in all_data.groupby('States')]

plt.bar(States, results['Sales'])
plt.xticks(States, rotation = 'vertical', size =8)
plt.ylabel('Sales inUSD ($)')
plt.xlabel('State Name')
plt.show()
```

### what time should we display ads to maximize likelihood of customer's buying the product?

```python
all_data['Order Date'] = pd.to_datetime(all_data['Order Date'])
```

```python
all_data.head()
```

## what products are usaully sold together?

```python
all_data.head()
```

```python
# we check the order ID

df = all_data[all_data['Order ID'].duplicated(keep=False)]
df.head(20)

df['Grouped'] = df.groupby('Order ID')['Product'].transform(lambda x: ','.join(x))
```

```python
df
```

```python
df = df[['Order ID','Grouped']].drop_duplicates()
```

```python
df.head()
```

```python
from itertools import combinations
from collections import Counter

count = Counter()

for row in df['Grouped']:
    row_list = row.split(',')
    count.update(Counter(combinations(row_list, 2)))
    
    
count.most_common(10)

```

## what product is sold the most and why do you think so?


```python
all_data.head()
```

```python
product_group = all_data.groupby('Product')

quantity_ordered = product_group.sum()['Quantity Ordered']

products = [product for product, df in product_group]


plt.bar(products,quantity_ordered)
plt.ylabel('Quantity Ordered')
plt.xlabel('Product')
plt.xticks(products, rotation = 'vertical', size =8)
plt.show()
```

```python
# hence the AAA batteries are sold the most. Simply because the cheap and more accessible.
```

```python
#to check the correlation with prices to test the reason given we compare with the graph of price each
prices= all_data.groupby('Product').mean()['Price Each']
print(prices)
```

```python
fig, ax1 = plt.subplots()

ax2 = ax1.twinx()
ax1.bar(products, quantity_ordered, color = 'g')
ax2.plot(products, prices, 'b-')

ax1.set_xlabel('Product Name')
ax1.set_ylabel('Quantity Ordered', color = 'g')
ax2.set_ylabel('Price ($)', color = 'b')
ax1.set_xticklabels(products, rotation ='vertical', size=8)

plt.show()
```

```python
# we see direct inverse relation with the price to the quantity bought. This may not be the generalization to all cases. 
```
